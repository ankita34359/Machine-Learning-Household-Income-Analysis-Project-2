```python
# Importing pandas as pd and numpy as np

import pandas as pd
import numpy as np
```

```python
# Loading the dataset into a pandas DataFrame

df = pd.read_csv("household_income_data.csv")
```

```python
# Step:1 Checking NaN values

df.isna().sum()
```

```python
# There is no NaN value so we don't need to fill the NaN value
```

```python
# Encoding of Categorical Data 
```

```python
# 1.) Using Map: We use map function for encoding the Sex and Homeownership_Status columns.

df['Gender'] = df['Gender'].map({"Male":0, "Female":1})
df['Homeownership_Status'] = df['Homeownership_Status'].map({"Own":0, "Rent":1})
```

```python
df.head()
```

```python
# 2.) Using Label Encode: We use LabelEncoder for Education_Level, Occupation, Location, Marital_Status, Employment_Status, Type_of_Housing and Primary_Mode_of_Transportation columns.

from sklearn.preprocessing import LabelEncoder
```

```python
df["Education_Level"] = LabelEncoder().fit_transform(df["Education_Level"])
df["Occupation"] = LabelEncoder().fit_transform(df["Occupation"])
df["Location"] = LabelEncoder().fit_transform(df["Location"])
df["Marital_Status"] = LabelEncoder().fit_transform(df["Marital_Status"])
df["Employment_Status"] = LabelEncoder().fit_transform(df["Employment_Status"])
df["Type_of_Housing"] = LabelEncoder().fit_transform(df["Type_of_Housing"])
df["Primary_Mode_of_Transportation"] = LabelEncoder().fit_transform(df["Primary_Mode_of_Transportation"])
```

```python
df.head()
```

```python
# Normalization

# We will normalize the column Age, Number_of_Dependents, Work_Experience, Income and Household_Size in the range of 0 to 1.
```

```python
# Importing MinMaxScaler for data preprocessing

from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalizing the Age, Number_of_Dependents, Work_Experience, Income and Household_Size columns.

df["Age"] = MinMaxScaler().fit_transform(df[["Age"]])
df["Number_of_Dependents"] = MinMaxScaler().fit_transform(df[["Number_of_Dependents"]])
df["Work_Experience"] = MinMaxScaler().fit_transform(df[["Work_Experience"]])
df["Household_Size"] = MinMaxScaler().fit_transform(df[["Household_Size"]])
df["Income"] = MinMaxScaler().fit_transform(df[["Income"]])
```

```python
df.head()
```

```python
# We will separate the Feature Varible and Target Variable

features = ["Age", "Education_Level", "Occupation", "Number_of_Dependents", "Location", "Work_Experience", "Employment_Status", "Household_Size", "Homeownership_Status", "Type_of_Housing", "Gender", "Primary_Mode_of_Transportation"]
X = df[features].values
Y = df["Income"].values
```

```python
# Displaying features columns 

df[features]
```

```python
# Get the features variable in numpy array

df[features].values
```

```python
# Now, Let's train the model

from sklearn.model_selection import train_test_split
```

```python
# Split the data in 70% and 30% for training and testing the model

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.03)
```

```python
# Algorithms abbrevations full form

# LR: Linear Regression
# LOR: Logistic Regression
# KNN: K-Nearest Neighbour
# RF: Random Forest
# GB: Gradient Boosting
# XGB: Extreme Gradient Boosting
```

```python
# In this we are using the LR, KNN, RF, GB and XGB alogorithm for taining the model

# Linear Regression
from sklearn.linear_model import LinearRegression

# K-Nearest Neighbors (for regression)
from sklearn.neighbors import KNeighborsRegressor

# Random Forest (for regression)
from sklearn.ensemble import RandomForestRegressor

# Gradient Boosting (for regression)
from sklearn.ensemble import GradientBoostingRegressor

# Extreme Gradient Boosting
import xgboost as xgb
from xgboost import XGBRegressor
```

```python
# Here we are creating a model and training the model

model_LR = LinearRegression().fit(X_train, Y_train)
model_KNN = KNeighborsRegressor(n_neighbors=12).fit(X_train, Y_train)
model_RF = RandomForestRegressor().fit(X_train, Y_train)
model_GB = GradientBoostingRegressor(n_estimators=100, random_state=42).fit(X_train, Y_train)
model_XGB = xgb.XGBRegressor().fit(X_train, Y_train)
```

```python
# Predicting the Test set results

Y_pred_LR = model_LR.predict(X_test)
Y_pred_KNN = model_KNN.predict(X_test)
Y_pred_RF = model_RF.predict(X_test)
Y_pred_GB = model_GB.predict(X_test)
Y_pred_XGB = model_XGB.predict(X_test)
```

```python
# For Regression we perform MSE, RMSE and r2_score
# Importing mean_squared_error, r2_score metrics

from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Calculating the Mean Squared Error

mse_LR = mean_squared_error(Y_test, Y_pred_LR)
mse_KNN = mean_squared_error(Y_test, Y_pred_KNN)
mse_RF = mean_squared_error(Y_test, Y_pred_RF)
mse_GB = mean_squared_error(Y_test, Y_pred_GB)
mse_XGB = mean_squared_error(Y_test, Y_pred_XGB)
```

```python
print("\nMean Squared Error:")

print(f"Linear Regression: {mse_LR}")
print(f"KNN: {mse_KNN}")
print(f"Random Forest: {mse_RF}")
print(f"Gradient Boosting: {mse_GB}")
print(f"XGBoost: {mse_XGB}")
```

```python
# Calculating the R-squared Score

r2_LR = r2_score(Y_test, Y_pred_LR)
r2_KNN = r2_score(Y_test, Y_pred_KNN)
r2_RF = r2_score(Y_test, Y_pred_RF)
r2_GB = r2_score(Y_test, Y_pred_GB)
r2_XGB = r2_score(Y_test, Y_pred_XGB)
```

```python
print("\nR-squared Score:")

print(f"Linear Regression: {r2_LR}")
print(f"KNN: {r2_KNN}")
print(f"Random Forest: {r2_RF}")
print(f"Gradient Boosting: {r2_GB}")
print(f"XGBoost: {r2_XGB}")
```

```python
# Calculating the Root Mean Squared Error

rmse_LR = mse_LR**0.5
rmse_KNN = mse_KNN**0.5
rmse_RF = mse_RF**0.5
rmse_GB = mse_GB**0.5
rmse_XGB = mse_XGB**0.5
```

```python
print("\nRoot Mean Squared Error:")

print(f"Linear Regression: {rmse_LR}")
print(f"KNN: {rmse_KNN}")
print(f"Random Forest: {rmse_RF}")
print(f"Gradient Boosting: {rmse_GB}")
print(f"XGBoost: {rmse_XGB}")
```

```python
# Random Forest is best-performing model on the basis of Mean Squared Error (MSE), R-squared Score and Root Mean Squared Error (RMSE).

# MSE (lower is better): Random Forest has the lowest MSE.
# R-squared (higher is better): Random Forest has the highest R-squared score.
# RMSE (lower is better): Random Forest has the lowest RMSE.
```
